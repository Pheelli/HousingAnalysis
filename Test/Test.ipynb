```python
# 1. Let's define the function to put movies into buckets based on their rating
def rating_bucket(x):
    if x >= 8.0:
        return "great"
    elif x >= 7.0:
        return "good"
    elif x >= 6.0:
        return "average"
    else:
        return "bad"
```

```python

# 2. Let's apply the function    
movies_df["Rating_Category"] = movies_df["Rating"].apply(rating_bucket)
```
